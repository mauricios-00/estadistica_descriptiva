#### Parte analítica de la matriz de covarianza

Primero hay que escalar los elementos de dataset:

1- Hay que usar la función `StandarScaler()` importada de *sklearn*  invocándola desde una variable
2- con una nueva variable, invocar a la anterior que contiene la función de ineterés y aplicarle una segunda función para ajustar los datos escalados `fis_transform()` que tendrá los siguiente parámetros:
* El dataset en cuestión pero aclarando qué variables son las que se van a usar para la matriz de covarianza, es por eso que no se requiere la variable categórica y no se agregará.

* * para observar todas las variables del dataset se puede invocar al dataset para después usar la variable implicita `columns` ( sin usar paréntesis)
* * * y el resultado de esto será el índice con los nombres de cada una  de las variables del dataset

Una vez elegidas las variables que se usarán simplemente se copian y se pegan, tener en cuenta que se trabaja con un array de dos dimentsiones  "[ [ ] ]" y  al invocar el dataset deberían de colorarse las variables de esta forma

resultando todo de la siguiente forma:

`scaler = StandardScaler()`
`scaled = scaler.fit_transform(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])`






```
iris.columns

scaler = StandardScaler()
scaled = scaler.fit_transform(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
scaled
```

#### Transpuesta del arreglo
Lo que se observa es un array con los elementos ya  normalizados, la normalización que tiene en cuanta restar por el promedio  cada uno de los elentos y dividirlo por la desvición estándar (sigma) 


Sin embargo los datos tal como están ordenados no sirven para alcular la matriz de correlaciones, por lo que hay que hacer una "transpuesta" del array. Para ello hay que aplicar a la variable escalada  el keyword `T`

`scaled.T`

El resultado son los mismo elementos pero ordenados de manera distinta, ahora todos los que están primero corresponden a los elementos de la primera variable encerrados en una [ ] , después de la segunda encerrados en una, etc

*El resultado esl a transpuesta de los datos escalados y normalizados*

y ahora ya es posible calcular la matriz de covarianza

```
scaled.T
```

Calcular la matriz de covarianza

1- definir una variable que contenga  `cov()` que es un método  de numpy, con parámetros:
* La transpuesta de los datos (*scaled.T*) 

2

```
covarianze_matrix = np.cov(scaled.T)
covarianze_matrix
```

Una manera de visualizar los datos de la matriz de covarianza es a través de un mapa de calor

1- invovar a la función `figure` de la librería matplotlib, con parámetro: `figsize= #,#` para determinar el tamaño 

2- invocar a la función `set` de la librería *seaborn*  con parámetros: `font_scale= ` para determinar el tamañño de la letra

3- comenzar a graficar el mapa de calor, para ello hay que invocar a la función `heatmap()` *seaborn* dentro de una nueva variable tendrá los siguientes parámetros:

* la variable que contiene el cálculo de la matriz de covarianza
* `cbar = True`  : mostrar escala de colores
* `annot = True` : que muestre los valores en cada elemento de la matriz
* `square = True` : que la matriz la muestre cuadrada
* `fmt = '.2f'` : formato para el tamaño de letra de los números, variables de la matriz
* `annot_kws = {}` con subparámetro `'size': 12` : otro  parámetro para el tamaño de letras de las variables de la matriz
* yticklabels = ['variable1', 'variable2', ..] : descripción de las etiquetas que deberían tener cada columna de la matriz
* xticklabels = ['variable1' , 'variable2' , ..] : descripción de las etiquetas de cada línea

```
plt.figure(figsize=(10,10))

sns.set(font_scale=1.5)

hm = sns.heatmap(covarianze_matrix,
                cbar = True,
                annot = True,
                square= True,
                fmt='.6f',
                annot_kws= {'size' :12},
                yticklabels= ['sepal_length', 'sepal_width' , 'petal_length' , 'petal_width'],
                xticklabels= ['sepal_length', 'sepal_width', 'petal_length' , 'petal_width'] )
```
